```python
import csv
from BeautifulSoup import BeautifulSoup as bs
import scipy.io as sio # for savemat and loadmat
from datetime import datetime
```

```python
data = None
with open('stocks.csv','r') as infile:
    csvin = csv.reader(infile)
    data = map(tuple,csvin)
    
# data[len(data)-1] # csv reader is a generator, so the data won't be available until we actually access it
```

```python
headers = data[0]
keys = [headers[0]]
keys += list(headers[2:6])
keys
```

```python
data_by_dates = dict()
for d in data[1:]:
    item = dict(zip(keys, (d[0],d[2],d[3],d[4],d[5])))
    if data_by_dates.has_key(d[1]):
        data_by_dates[d[1]].append( item )
        # data_by_dates[datetime.strptime(d[1],'%d-%b-%y').date()].append(item)
    else:
        data_by_dates[d[1]] = [ item ]
        # data_by_dates[datetime.strptime(d[1],'%d-%b-%y').date()] = [ item ]
```
